```python
import pandas as pd
import numpy as np
import matplotlib as py
```

```python
data0 = pd.read_csv("daily_sales_data_0.csv")
data1 = pd.read_csv("daily_sales_data_1.csv")
data2 = pd.read_csv("daily_sales_data_2.csv")
```

# Analysis of first csv files

```python
data0.head()
```

```python
data0.isnull().sum()
```

```python
data0 = data0[data0.apply(lambda row: row.astype(str).str.contains('pink morsel', case=False).any(), axis=1)]
```

```python
data0.head()
```

```python
data0['price'] = data0['price'].astype(str).str.replace(r'[\$,]', '', regex=True).str.strip()
data0['price'] = pd.to_numeric(data0['price'], errors='coerce')
```

```python
data0.head()
```

```python
data0['Sales'] = data0['quantity'] * data0['price']
data0.drop(columns=['quantity','price'], inplace=True)
```

```python
data0.head()
```

# Analysis of second csv files

```python
data1.head()
```

```python
data1.isnull().sum()
```

```python
data1 = data1[data1.apply(lambda row:row.astype(str).str.contains('pink morsel', case=False).any(), axis=1)]
```

```python
data1.head()
```

```python
data1['price'] = data1['price'].astype(str).str.replace(r'[\$,]', '', regex=True).str.strip()
data1['price'] = pd.to_numeric(data1['price'], errors='coerce')
data1.head()
```

```python
data1['Sales'] = data1['quantity'] * data1['price']
data1.drop(columns=['quantity','price'], inplace=True)
data1.head()
```

# Analysis of third files
